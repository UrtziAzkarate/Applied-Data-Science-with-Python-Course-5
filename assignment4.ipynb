```python
def salary_predictions():
    from sklearn.preprocessing import StandardScaler
    from sklearn.ensemble import RandomForestClassifier
    from sklearn.model_selection import train_test_split
    from sklearn.linear_model import LinearRegression, LogisticRegression
    from sklearn.svm import SVC
    from sklearn.dummy import DummyClassifier
    from sklearn.metrics import precision_recall_curve, roc_curve, roc_auc_score

    df1 = pd.DataFrame(G.nodes(data=True))
    df1['Dep'] = df1[1].apply(lambda x: x['Department'])
    df1['Management'] = df1[1].apply(lambda x: x['ManagementSalary'])
    df1.drop(columns=[1], inplace=True)
    df1.rename(columns={0:'Node'}, inplace=True)
    
    test_data = df1[df1['Management'].isna()].drop(columns=('Management'))
    
    training_data = df1[~df1['Management'].isna()]
    training_data_X = training_data.drop(columns='Management')
    training_data_y = training_data['Management']

    
    X_train, X_test, y_train, y_test = train_test_split(training_data_X, training_data_y, random_state=0)
    
    
    #svc = SVC(C=1, gamma=10).fit(X_train, y_train)
    #dummy = DummyClassifier(strategy='most_frequent').fit(X_train, y_train)

    logreg = LogisticRegression().fit(X_train, y_train)
    
    pred = logreg.predict_proba(test_data)

    resultdf = test_data.set_index('Node').drop(columns=('Dep'))
    resultdf['pred'] = pred[:,1] #just consider the second column, the prob of the result being 1 (positive)
    
    #roc_auc = roc_auc_score(pred, y_test)  # Provide true labels (y_true) for test data
    #pred = logreg.predict(X_test)

    return resultdf.squeeze() #logreg.score(X_test, y_test)

salary_predictions()
```

```python
ans_salary_preds = salary_predictions()
assert type(ans_salary_preds) == pd.core.series.Series, "You must return a Pandas series"
assert len(ans_salary_preds) == 252, "The series must be of length 252"
```

```python
def new_connections_predictions():
    from sklearn.preprocessing import StandardScaler, LabelEncoder
    from sklearn.ensemble import RandomForestClassifier
    from operator import itemgetter
    from sklearn.model_selection import train_test_split
    from sklearn.linear_model import LinearRegression, LogisticRegression
    from sklearn.svm import SVC
    from sklearn.dummy import DummyClassifier
    from sklearn.metrics import precision_recall_curve, roc_curve, roc_auc_score
    
    
    df2 = future_connections.copy()
    
    df2['send'] = list(map(itemgetter(0),df2.index))
    df2['receive'] = list(map(itemgetter(1),df2.index))
    df2['fut_connection'] = df2['Future Connection']
    df2.drop(columns=('Future Connection'), inplace=True)
    
    t = pd.DataFrame(nx.degree(G)).rename(columns={0:'node', 1:'degree'})
    t['closeness_c'] = nx.closeness_centrality(G).values()
    t['betweenness_c'] = nx.betweenness_centrality(G).values()
    t['degree_c'] = nx.degree_centrality(G).values()

    #nx.shortest_path_length(G, source=6, target=840)
    
    test = df2.merge(t, left_on='send', right_on='node', how='left')
    test.rename(columns={'degree': 's_degree', 'closeness_c':'s_closeness_c', 'betweenness_c':'s_betweenness_c', 'degree_c':'s_degree_c'}, inplace=True)
    test2 = test.merge(t, left_on='receive', right_on='node', how='left')
    
    finaldf = test2.drop(columns=['node_x', 'node_y'])
    
    trainingdf = finaldf[~finaldf['fut_connection'].isna()]
    testingdf = finaldf[finaldf['fut_connection'].isna()].drop(columns=('fut_connection'))
    
    trainingdf_X = trainingdf[['send', 'receive', 's_degree', 's_closeness_c', 's_betweenness_c', 's_degree_c', 'degree', 'closeness_c', 'betweenness_c', 'degree_c']]
    trainingdf_y = trainingdf['fut_connection']
    
    X_train, X_test, y_train, y_test = train_test_split(trainingdf_X, trainingdf_y, random_state=0)
    
    logreg = RandomForestClassifier().fit(X_train, y_train)
    score = logreg.score(X_test, y_test)
    
    pred = logreg.predict_proba(testingdf)[:,1]
    
    testingdf['prediction'] = pred
    testingdf['new_index'] = testingdf.apply(lambda row: (round(row['send']), round(row['receive'])), axis=1)
        
    return testingdf.set_index('new_index')['prediction'].squeeze()

new_connections_predictions()
```

```python
ans_prob_preds = new_connections_predictions()
assert type(ans_prob_preds) == pd.core.series.Series, "You must return a Pandas series"
assert len(ans_prob_preds) == 122112, "The series must be of length 122112"
```
